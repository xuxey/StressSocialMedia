# ANN

## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Combining training files

```python
dataset_train = pd.read_csv(r'data/reddit-train.csv')
dataset_test = pd.read_csv('data/reddit-test.csv')
frames = [dataset_train, dataset_test]
dataset = pd.concat(frames)
```

## Isolating Independent & Dependent Variables

```python
X = dataset.iloc[:, 7:]
subreddits = np.array(dataset.iloc[:, 0])
X['subreddits'] = subreddits
y = dataset.iloc[:, dataset.columns == 'label'].values
```

## Encoding subreddits

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

## Splitting into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## Training Artificial Neural Network

Inititalizing the ANN

```python
import tensorflow as tf
ann = tf.keras.models.Sequential()
```

Adding First and Second Hidden Layers

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

Adding output layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

Compiling and training the ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
ann.fit(X_train, y_train, batch_size = 32, epochs = 150)
```

## Predicting Test Set results

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Check Accuracy using Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
